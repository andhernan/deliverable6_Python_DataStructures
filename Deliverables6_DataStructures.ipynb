```python
#step 1: import libraries needed
import pandas as pd

import numpy as np

import seaborn as sns
```

```python
#step 2: load the "iris" dataset and store it as a data frame.

iris = sns.load_dataset('iris')

#show me it
iris.head()
```

```python
#step 3: delete the species column (it is unnecessary and repetitive)
iris.drop(labels = 'species', axis = 1, inplace = True)
#1 is drop columns
#0 is drop rows

#show me it
iris.head()
```

```python
#Step 4: Add two more columns to the iris data frame, "sepal_sum" and "petal_sum". 
#The "sepal_sum" is the sum of the values from the "sepal_length" and "sepal_width" columns. 
#On the other hand, The "petal_sum" is the sum of the values from the "petal_length" and "petal_width" columns.

#First sepal_sum
iris['sepal_sum'] = iris['sepal_length'] + iris['sepal_width']

#Second petal_Sum
iris['petal_sum'] = iris['petal_length'] + iris['petal_width']

#now THEORETICALLY that should work
#show me the MAGIC
iris.head()
```

```python
#step 5: Create a new data frame with the summary stats for the iris dataset. 
#For this summary, compute each column's mean, standard deviation, and minimum and maximum values.

mean = list(iris.mean())
stdev = list(iris.std())
min = list(iris.min())
max = list(iris.max())
summary_dict = {'mean':mean,'stdev':stdev,'min':min,'max':max}
index = ['sepal_length','sepal_width','petal_length','petal_width','sepal_sum', 'petal_sum']

summary_df = pd.DataFrame(data = summary_dict, index = index)
summary_df
```
